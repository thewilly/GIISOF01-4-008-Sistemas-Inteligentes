A continuación definiremos nuestro banco de datos que posteriormente usaremos en los modelos de aprendizaje. El banco de datos que usaremos se corresponde con expresiones genéticas de pacientes con distintos tipo de tumores: BRCA, KIRC, COAD, LUAD and PRAD. [Aquí](https://archive.ics.uci.edu/ml/datasets/gene+expression+cancer+RNA-Seq) encontrarás más información sobre el banco de prueba.

```r
# Carga de los datos en variables.
geneLevel <- read.csv('data/data.csv',stringsAsFactors = F)
label <- read.csv('data/labels.csv',stringsAsFactors = F)

# Juntamos ambos archivos en una tabla.
tcga <- merge(geneLevel, label, sort = F)
tcga$X <- NULL
```

Los datos anteriores contienen 20531 variables y 801 instancias, debido a que nuestro ordenador no tiene memoria suficiente para un problema tan grande seleccionaremos sólo 100 variables, que serán elegidas según mi identificador universitario.

```r
# Filtramos la tabla de datos a 100 variables según
# mi identificador universitario, UO23
set.seed(236856) # 236856 es mi identificador de la Universidsad de Oviedo.
tcga.filtered <- tcga[,c(sample(ncol(tcga)-1,100),ncol(tcga))]
```
