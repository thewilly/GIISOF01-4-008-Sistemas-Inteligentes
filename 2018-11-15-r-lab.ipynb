```r
 wifi <- read.delim( "https://archive.ics.uci.edu/ml/machine-learning-databases/00422/wifi_localization.txt", header = F)
```

```r
write.csv(wifi,"wifi.csv",row.names = F)
```

```r
summary(wifi)
```

```r
wifi$V8 <- as.factor(wifi$V8)
```

```r
library(ggplot2)
ggplot(wifi , aes(x=V1)) + geom_density()+facet_wrap(~V8)
```

```r
library(reshape2)
ggplot(melt(wifi) , aes(x=value,color=V8)) + geom_density()+facet_grid(V8~variable)
```

```r
head(melt(wifi))
```

```r
ggplot(wifi , aes(x=V1,y=V2,color=V8)) + geom_point()
```

```r
makePairs <- function(data,classIndex){
    dfm <- NULL
    for(i in 1:ncol(data)){
        if(i==classIndex)
            next
        for(j in 1:ncol(data)){
            if(j==classIndex)
                next
            if(i==j)
                next
            dfm <- rbind(dfm,data.frame(x=data[,i],y=data[,j],
            Var1=colnames(data)[i],
            Var2=colnames(data)[j],
            class=data[,classIndex]))
        }
    }
    dfm$Var2 <- relevel(dfm$Var2,levels(dfm$Var1)[1])
    dfm
}
dataToPlot <- makePairs(wifi,8)
ggplot(dataToPlot, aes_string(x = "x", y = "y")) + 
    facet_grid(Var1 ~ Var2, scales = "free") +
    geom_point(aes(colour=class), alpha=0.6)
```
